```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import csv
source1 = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source1,'lxml')
csv_file = open('TorontoNeighborhood_scrape.csv','w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Postalcode','Borough', 'Neighborhood'])
df = pd.read_html(source1, header=0, attrs={"class":"wikitable sortable"})[0]
df.groupby(['Postcode','Borough'], sort = False).agg(lambda X:','.join(X))
```
